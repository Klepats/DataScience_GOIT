## Створіть нову таблицю python_data, в якій будуть тільки рядки зі спеціалістами, які вказали мову програмування Python

```python
python_data=pd.DataFrame()
python_data=pd.read_csv('2017_jun_final - 2017_jun_final.csv')
python_data = python_data[python_data['Мова програмування'] == "Python"]
python_data
```

## Визначення розміру таблиці python_data за допомогою методу shape

```python
python_data.shape
```

## Використовуючи метод groupby, виконайте групування за стовпчиком "Посада"

```python
grouped_by_position = python_data.groupby('Посада')
```

```python
aggregated_data = grouped_by_position.agg({'Зарплата на місяць': ['min', 'max']})

aggregated_data.columns = ['Мінімальна_зарплата', 'Максимальна_зарплата']

aggregated_data
```

## Створіть функцію fill_avg_salary, яка повертатиме середнє значення заробітної плати на місяць. Використовуйте її для методу apply та створіть новий стовпчик "avg"

```python
def fill_avg_salary(salary_column):
    return (salary_column.min() + salary_column.max()) / 2

aggregated_data['avg'] = aggregated_data.apply(fill_avg_salary, axis=1)

aggregated_data.sort_values('avg', ascending=False)
```

## Створіть описову статистику за допомогою методу describe для нового стовпчика.

```python
aggregated_data['avg'].describe()
```

## Збережіть отриману таблицю в CSV файл

```python
aggregated_data.to_csv("aggregated_data.csv", index=False)
```

## Графіки

```python
sns.boxenplot(x='Місто', y='Зарплата на місяць', data=df)
plt.xlabel('Місто')
plt.ylabel('Зарплата в місяць')
plt.title('Росподілення зарплат по містам')
plt.xticks(rotation=90)
plt.show()
```

```python
sns.scatterplot(x='Загальний досвід роботи', y='Зміна зарплати за 12 місяців', data=df)
plt.xlabel('Загальний досвід работы (років)')
plt.ylabel('Зміна зарплати за 12 місяців')
plt.title('Залежність зміни зарплати від досвіду роботи')
plt.show()
```

```python
sns.histplot(data=df, x='Зарплата на місяць', kde=True)
plt.xlabel('Зарплата в місяць')
plt.ylabel('Частота')
plt.title('Росподіл зарплат за місяць')
plt.show()
```
